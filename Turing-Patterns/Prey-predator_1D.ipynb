```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
D1=1
D2=1
a=1
r=1
N0=1
e=1
lmda=1
A=1
K=1
m=1

size = 100 
dx = 100. / size  
T = 1.0
dt = 0.01
T=(1/(r*K))*T
n = int(T / dt)
```

```python
alpha=(a*r)/(lmda**2)
beta=N0/K
sigma=m/(K**2 *e*lmda)
h=A*K**2*lmda
eta=(K*e*lmda)/r
```

```python
N=10
P=1000
N_arr=[]
P_arr=[]
N_arr.append(N)
P_arr.append(P)
```

```python


for i in range(n):
    
    
    Nc = N
    Pc = P
    
    N, P = \
      Nc + dt * (Nc*(1-Nc)*(N-beta)-(1-alpha*Pc)*(Nc**2*P)/(1+h*(1+alpha*Pc)*N**2)), \
        Pc + dt * eta*((1+alpha*Pc)*Nc**2*Pc/(1+h*(1+alpha**2*Pc)*Nc**2)-sigma*Pc)
    N_arr.append(N)
    P_arr.append(P)
    
```

```python
N_new=N*K
P_new=P*(r/lmda)
```

```python
time_values = np.arange(0, len(N_arr) * dt, dt)

plt.figure(figsize=(10, 6))
plt.plot(time_values, P_arr, label='Predator', color='b')
plt.plot(time_values, N_arr, label='Prey', color='y')
plt.xlabel('Time (dt)')
plt.ylabel('Population')
plt.title('Prey-Predator Populations vs. Time')

plt.legend()
plt.grid(True)

plt.show()
```
