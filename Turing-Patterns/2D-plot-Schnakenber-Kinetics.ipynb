```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import networkx as nx
from matplotlib.animation import FuncAnimation, PillowWriter 
```

```python
D1 = 1.0
D2 = 40.0
a = 0.1
b = 0.9
size = 100  # size of the 2D grid

T = 200
dt = 0.01
dx=(dt/0.005)**0.5
n = int(T / dt)
```

```python
U = np.random.rand(size, size)
V = np.random.rand(size, size)
```

```python
def laplacian2D(a, dx):
    return (
        - 4 * a
        + np.roll(a,1,axis=0) 
        + np.roll(a,-1,axis=0)
        + np.roll(a,+1,axis=1)
        + np.roll(a,-1,axis=1)
    ) /(dx*dx)
```

```python
step_plot = n // 9
for i in range(n):
    # Compute Laplacians
    deltaU = laplacian2D(U, dx)
    deltaV = laplacian2D(V, dx)
    
    Uc = U
    Vc = V
    
    # Update concentrations using Schnakenberg kinetics
    U = \
       Uc + dt * (D1 * deltaU + a - Uc + (Uc**2) * Vc)
    V = \
      Vc + dt * (D2 * deltaV + b -( Uc**2) * Vc)
    
 

    if i % step_plot == 0 and i < 9 * step_plot:
      plt.imshow(U, cmap='jet', interpolation='nearest')
  
      plt.title(' concentration of U')
      plt.colorbar()
      plt.show()








   
```
