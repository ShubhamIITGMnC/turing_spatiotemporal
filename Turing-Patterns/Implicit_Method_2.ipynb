```python
 import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve

# Parameters
L = 10.0  # Spatial domain length
T = 1.0   # Total time
Nx = 100  # Number of spatial grid points
Nt = 100  # Number of time steps
alpha = 0.01  # Time step size
h = L / Nx  # Spatial step size

# Create spatial grid
x = np.linspace(0, L, Nx+1)

# Create time grid
t = np.linspace(0, T, Nt+1)

# Initialize solution matrix
U = np.zeros((Nx+1, Nt+1))

# Set initial condition
U[:, 0] = np.sin(np.pi * x)

# Construct the tridiagonal matrix for the implicit method



# Time-stepping loop
for n in range(0, Nt):
    A = np.eye(Nx+1) * (1 + 2*alpha-alpha*h*h*U[:, n]) - np.eye(Nx+1, k=1) * alpha - np.eye(Nx+1, k=-1) * alpha
    b = U[:, n] 
    U[:, n+1] = solve(A, b)  # Solve the linear system using scipy's solve


plt.figure(figsize=(10, 6))
plt.plot(x, U[:, -1], label='Final Condition')
plt.xlabel('Space')
plt.ylabel('U')
plt.title('Final Condition of PDE: U_t = U_xx + U^2')
plt.legend()
plt.grid()
plt.show()

```
