```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

```python
L = 10.0  # Spatial domain length
T = 1.0   # Total time
Nx = 1000  # Number of spatial grid points
Nt = 100  # Number of time steps
alpha = 0.01  # Time step size
h = L / Nx  # Spatial step size
```

```python
# Create spatial grid
x = np.linspace(0, L, Nx+1)

# Create time grid
t = np.linspace(0, T, Nt+1)

# Initialize solution matrix
U = np.zeros((Nx+1, Nt+1))

# Set initial condition
U[:, 0] = np.sin(np.pi * x)
```

```python
A = np.eye(Nx+1) * (1 + 2*alpha) - np.eye(Nx+1, k=1) * alpha - np.eye(Nx+1, k=-1) * alpha
```

```python
for n in range(0, Nt):
    b = U[:, n] + alpha * U[:, n]**2*h*h  # Right-hand side
    U[:, n+1] = solve(A, b)  # Solve the linear system using scipy's solve

```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, U[:, -1], label='Final Condition')
plt.xlabel('Space')
plt.ylabel('U')
plt.title('Final Condition of PDE: U_t = U_xx + U^2')
plt.legend()
plt.grid()
plt.show()




```
